```python
def df_split(df):
  categoria = []
  for x in df['category_name']:
    categoria.append(x.split('/'))
  return categoria


def sub_categoria(df):  
  lista = df_split(df)
  for y in range(len(lista)):
    lista[y] = lista[y][1]
  return lista
```

```python
filtro_other = pd.read_csv('https://raw.githubusercontent.com/AugustoCRX/bootcamp_blue/main/novos_arquivos_treino/Limpeza%20v3/filtro_Other_train_v3.csv')
```

```python
filtro_other.head()
```

```python
filtro_other['category_name'] = filtro_other['category_name'].replace('Other','Other/Other')
```

```python
filtro_other['sub_category_name'] = sub_categoria(filtro_other)
```

```python
filtro_other.sub_category_name.value_counts()
```

```python
import statsmodels.api as sm
```

```python
filtro_other = pd.get_dummies(filtro_other, columns=['sub_category_name','brands_filter'])
```

```python
X = filtro_other.drop(['name','category_name','item_description','date','stock','date2','day','month','year','stock','price','brand_name','description_clear','name_clear'],axis=1)
y = filtro_other.price
X = sm.add_constant(X)
rl = sm.OLS(exog=X,endog=y)
result = rl.fit()
print(result.summary())
```

```python
ypred = result.predict(X)
```

```python
from statsmodels.tools.eval_measures import rmse

RMSE = rmse(y,ypred)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
dtr = DecisionTreeRegressor(random_state=10)
result2 = dtr.fit(X2,y2)
```

```python
R2 = result2.score(X2,y2)
```

```python
R2
```

```python
y_pred1 = result2.predict(X2)
```

```python
mean_squared_error(y2,y_pred1)
```

```python
result3 = dtr2.fit(X2,y2)
```
