# Modelo Sport

```python
def criar_tabela_sport():
  sport = pd.DataFrame(columns=['item_condition_id', 'sub_category_name_Apparel', 
  'sub_category_name_Exercise', 'sub_category_name_Fan Shop', 'sub_category_name_Footwear', 
  'sub_category_name_Golf', 'sub_category_name_Other', 'sub_category_name_Outdoors', 
  'sub_category_name_Team Sports', 'brands_filter_adidas', 'brands_filter_basics_brand',
  'brands_filter_burton', 'brands_filter_callaway', 'brands_filter_capezio', 
  'brands_filter_champion', 'brands_filter_easton', 'brands_filter_everlast', 
  'brands_filter_fitbit', 'brands_filter_garmin', 'brands_filter_intermediary_brand', 
  'brands_filter_luxury_brand', 'brands_filter_mizuno', 'brands_filter_nike',
  'brands_filter_popular_brand', 'brands_filter_rawlings', 'brands_filter_reebok',
  'brands_filter_taylormade', 'brands_filter_titleist', 'brands_filter_under armour', 
  'brands_filter_upper_popular_brand', 'brands_filter_wilson', 'brands_filter_yeti', 
  'shipping_0', 'shipping_1'])

  sport.loc[0] = 0
  return sport
```

```python
def listar_brands_sport():
  lista = pd.read_csv('Dados/Brands_name_sports _ outdoors.csv')
  brands_sport = lista['brands']
  brands_sport = [x for x in brands_sport if pd.isnull(x) == False]
  basics_brand_sport = lista['basics_brand']
  basics_brand_sport = [x for x in basics_brand_sport if pd.isnull(x) == False]
  popular_brand_sport = lista['popular_brand']
  popular_brand_sport = [x for x in popular_brand_sport if pd.isnull(x) == False]
  upper_pop_brand_sport = lista['upper_pop_brand']
  upper_pop_brand_sport = [x for x in upper_pop_brand_sport if pd.isnull(x) == False]
  intermediary_brand_sport = lista['intermediary_brand']
  intermediary_brand_sport = [x for x in intermediary_brand_sport if pd.isnull(x) == False]
  return brands_sport, basics_brand_sport, popular_brand_sport, upper_pop_brand_sport, intermediary_brand_sport
```

```python
def verificar_brand_sport():
  marca = input("Digite marca: ").strip().lower()
  brands, basics_brand, popular_brand, upper_pop_brand, intermediary_brand = listar_brands_sport()
  concatenacao_marca = 'brands_filter_'
  if marca in brands:
    concatenacao_marca += marca
    return concatenacao_marca
  elif marca in basics_brand:
    concatenacao_marca += 'basics_brand'
    return concatenacao_marca
  elif marca in popular_brand:
    concatenacao_marca += 'popular_brand'
    return concatenacao_marca
  elif marca in upper_pop_brand:
    concatenacao_marca += 'upper_popular_brand'
    return concatenacao_marca
  elif marca in intermediary_brand:
    concatenacao_marca += 'intermediary_brand'
    return concatenacao_marca
  else:
    print('Marca não identificada')
    classificacao = int(input('Classifique sua marca entre 1 e 5, onde 1 marca basica e 5 marca luxuosa:' ))
    if classificacao == 1:
      concatenacao_marca += 'basics_brand'
      return concatenacao_marca
    elif classificacao == 2:
      concatenacao_marca += 'popular_brand'
      return concatenacao_marca
    elif classificacao == 3:
      concatenacao_marca += 'upper_popular_brand'
      return concatenacao_marca
    elif classificacao == 4:
      concatenacao_marca += 'intermediary_brand'
      return concatenacao_marca
    else:
      concatenacao_marca += 'luxury_brand'
      return concatenacao_marca
```

```python
def coletar_subcategoria_sport():
  concatenacao_categoria = 'sub_category_name_'
  print('''Lista de SubCategorias
  Apparel
  Exercise
  Fan Shop 
  Footwear
  Golf 
  Other
  Outdoors
  Team Sports''')
  sub_categoria = input('Digite a Subcategoria: ').strip().title()
  concatenacao_categoria += sub_categoria
  return concatenacao_categoria
```

```python
def coletar_shipping_sport():
  concatenacao_shipping = 'shipping_'
  print(''' Frete
  0 - comprador
  1 - vendedor''')
  shipping = input('Digite o número conforme legenda para quem será responsavel pelo custeio do frete: ')
  concatenacao_shipping += shipping
  return concatenacao_shipping
```

```python
def rodar_modelo_sport(df):
  modelo = open('Modelos/modelo_sport','rb')
  lm_new = pickle.load(modelo)
  modelo.close()
  valor = lm_new.predict(df).round(2)
  print(f'Preço sugerido = {valor[0]}')
```

```python
def rodar_sport():
  dados_sport = criar_tabela_sport()
  dados_sport[coletar_subcategoria_sport()] = 1
  dados_sport[verificar_brand_sport()] = 1
  dados_sport[coletar_shipping_sport()] = 1
  item_condition = int(input("'Digite condição do item entre 1 e 5, onde 1 - Excelente e 5 - Péssimo: '"))
  dados_sport['item_condition_id'] = item_condition
  rodar_modelo_sport(dados_sport)
  print('')
```
